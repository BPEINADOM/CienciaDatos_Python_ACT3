# Proceso ETL con Pandas

## Objetivo
Realizar limpieza y transformación de datos desde un archivo Excel utilizando Pandas.

```python
# Importamos la libreria Pandas
import pandas as pd
```

## 1. Quitar columnas que no se necesiten

```python
# Cargamos el Excel ignorando las primeras 6 filas
df = pd.read_excel('../data/raw/Datos_Reto.xlsx', skiprows=6)

# Eliminar columna innecesaria
df = df.drop(columns=["Unnamed: 0"])
```

## 2. Quitar filas que no se necesiten

```python
# Eliminamos filas innecesarias haciendo uso de "!=" para que se quede con todas las filas excepto las que tengan esos textos exactos.
df = df[
    (df["Área"] != "Total empleados") &
    (df["Área"] != "Total área:") &
    (df["Área"] != "Área")
]
```

## 3. Garantizar el correcto nombre de los campos

```python
# Corregimos nombre de la columna
df = df.rename(columns={
    "Títul profesional": "Título profesional"
})
```

## 4. Garantizar que todos los registros tengan el código de área

```python
# Rellenamos hacia abajo los espacios de las celdas combinadas
df["Área"] = df["Área"].ffill()
```
